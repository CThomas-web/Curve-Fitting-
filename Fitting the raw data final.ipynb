```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
```

```python
def err(a,x,y):
    return f(a,x) - y
```

Caesium137 - Fitting the Photopeak

```python
caesium = np.loadtxt('caesium137.csv', delimiter = ',,', skiprows = 200)
caesium_chan_fit = caesium[0:150,0]
caesium_count_fit = caesium[0:150,1]
```

```python
def f(a,x):
    return (a[0]*np.exp(-a[1]*(x-a[2]))) + a[3] + (a[4]*np.exp(-((x-a[5])**2)/(2*(a[6]**2))))
a0 = [5000,0.01,0.01,10,10,260,1]

a, success = leastsq(err, a0, args = (caesium_chan_fit,caesium_count_fit))
#plotting the fit
plt.figure(figsize=(10,5))
x = np.linspace(caesium_chan_fit.min(), caesium_chan_fit.max(), 1000)
plt.plot(caesium_chan_fit, caesium_count_fit,'b.',label = 'Caesium137-photopeak-slice ')
plt.plot(x, f(a, x), 'r-', label = 'fit: least squares minimization')
plt.xlabel('channel',size = 14)
plt.ylabel('counts',size = 14)
plt.title('Caesium-137 Photopeak_Fit', size = 16)
plt.legend()
plt.show()
```
